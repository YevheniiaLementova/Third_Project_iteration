```python
import pandas as pd
```

## Prepare first file

```python
data = pd.read_json('combined-convo.json')
```

```python
data 
```

```python
frst_part_test = data.loc[data['id'].isin([13, 21, 825, 833, 1223])]
```

```python
frst_part_test
```

```python
data.drop(data[data['id'].isin([13, 21, 825, 833, 1223])].index, inplace=True)
```

## Prepare second file

```python
data2 = pd.read_json('combined-convo_2.json')
```

```python
data2
```

```python
sec_part_test = data2.loc[data2['id'].isin([16, 23, 852, 855, 1177])]
```

```python
sec_part_test
```

```python
data2.drop(data2[data2['id'].isin([16, 23, 852, 855, 1177])].index, inplace=True)
```

## Prepare third file

```python
data3 = pd.read_json('combined-convo_3.json')
```

```python
data3
```

```python
third_part_test = data3.loc[data3['id'].isin([20, 25, 872, 883, 1474])]
```

```python
third_part_test
```

```python
data3.drop(data3[data3['id'].isin([20, 25, 872, 883, 1474])].index, inplace=True)
```

## Concatenate train dataset

```python
train_dataset = pd.concat([data, data2, data3], ignore_index=True)
```

```python
train_dataset
```

```python
train_dataset.drop('id', axis = 1, inplace = True)
```

```python
train_dataset
```

## Concatenate test dataset

```python
test_dataset = pd.concat([frst_part_test, sec_part_test, third_part_test], ignore_index=True)
```

```python
test_dataset
```

```python
test_dataset.drop('id', axis = 1, inplace = True)
```

```python
test_dataset
```

## Save datasets

```python
train_dataset.to_json('train_dataset.json', orient='records')
```

```python
test_dataset.to_json('test_dataset.json', orient='records')
```
